```python
def get_data(folder, filetype):
    # greyscaling the input images
    image_names = list(filter(lambda f: f.endswith(filetype), os.listdir(folder)))
    ims = [f'{folder}/{image_name}' for image_name in image_names]

    im_color = [cv2.imread(im) for im in ims]
    im_gray = [cv2.cvtColor(im, cv2.COLOR_BGR2GRAY) for im in im_color]
    
    # save the grayscale images
    gray_filenames = [f'{folder}/{image_name}'.replace(f'{filetype}', '_bw.png') for image_name in image_names]
    for idx, image in enumerate(im_gray):
        cv2.imwrite(gray_filenames[idx], im_gray[idx])

    # get text data from image
    texts = []
    for gray_path in gray_filenames:
        texts.append([pytesseract.image_to_string(Image.open(gray_path))])

    return texts
```

```python
texts = get_data('images/table_screenshots', '.png')
```
